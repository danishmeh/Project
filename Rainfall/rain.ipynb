```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv("Rainfall_1901_2016_PAK.csv")
df.head()
```

```python
# EDA 
# Remove spaces and dashes 
df.rename(columns={'Rainfall - (MM)':'Rainfall(mm)',' Year':'year'},inplace=True)
df.columns
```

## Rainfall occurance with respect to Month
## 1901 to 2016

```python
fig = px.bar(df, x='Month', y='Rainfall(mm)',color='Rainfall(mm)')
fig.update_layout(title_text="Rainfall from 1901 to 2016", title_x=0.5)
fig.update_layout(
    font_family="Courier New",
    font_color="blue",font_size=15,
    title_font_family="Times New Roman",
    title_font_color="red", yaxis_title="Rain occurance",
    legend_title_font_color="blue")
fig.add_vrect(x0="July", x1="August", 
              annotation_text="Flood Area", annotation_position="top left",
              fillcolor="green", opacity=0.50, line_width=1)
```

###  RainFall(mm) Avg of All Years of Months from 1901 to 2016
  July is the highest Avg rate of Rainfall is 56.64mm   
  August Rainfall(mm) Avg rate is 51.8mm  
  November is the minimum avg rate of Rainfall 5.8mm

```python
plt.figure(figsize=(15,5))
ax = plt.axes()
plt.title("RainFall(mm) Avg of All Month from 1901 to 2016",size=16)
plt.xlabel("Month",size=14)
plt.ylabel("Rainfall",size=14)
ax.xaxis.label.set_color('blue')          #setting up X-axis label color to yellow
ax.yaxis.label.set_color('blue')          #setting up Y-axis label color to blue
# ax.bar_label(ax.containers[0])
p = sns.barplot(data=df,x='Month',y='Rainfall(mm)',width=0.8,saturation=1, capsize=0.06,errwidth=0.1)
#show values on barplot
for container in p.containers:
    ax.bar_label(container)
    
```

## RainFall with Respect to year

```python
import plotly.express as px
plt.figure(figsize=(50,50))
fig = px.line(df, x='year', y="Rainfall(mm)",line_group='year')
fig.add_hrect(y0="100", y1="160", 
              annotation_text="Flood Alarm", annotation_position="top left",
              fillcolor="green", opacity=0.20, line_width=1)
fig.update_layout(
    font_family="Courier New",
    font_color="red",font_size=15,title="Flood Indication with Respect to year",
    title_font_family="Times New Roman",
    title_font_color="Red", yaxis_title="Rainfall(mm)",
    legend_title_font_color="blue")

fig.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(df['year'], df['Rainfall(mm)'])
plt.xlabel('Year')
plt.ylabel('RainFall(mm)')
plt.title('Rainfall vs year', y=1.1)
plt.grid()
plt.show()
```

## RainFall average with respect to years(1901 to 2016)

```python
plt.figure(figsize=(15,5))
df = df.groupby('year').mean()
plt.plot(df.index, df['Rainfall(mm)'])
plt.xlabel('year')
plt.ylabel('Rainfall(mm) in Avg')
plt.title('Rainfall vs year', y=1.1)
plt.grid()
plt.show()
```

## dot plot with respect to higest monthly points

```python
import plotly.io as pio

subject = df['Month']
score = df['Rainfall(mm)']

data = [dict(
  type = 'line',
  x = subject,
  y = score,
  mode = 'markers',
  transforms = [dict(
    type = 'aggregate',
    groups = subject,
    aggregations = [dict(
        target = 'y', func = 'max', enabled = True),
    ]
  )]
)]

fig_dict = dict(data=data)

pio.show(fig_dict, validate=False)
```

## dot plot with respect to lowest monthly points 

```python
import plotly.io as pio

subject = df['Month']
score = df['Rainfall(mm)']

data = [dict(
  type = 'line',
  x = subject,
  y = score,
  mode = 'markers',
  transforms = [dict(
    type = 'aggregate',
    groups = subject,
    aggregations = [dict(
        target = 'y', func = 'min', enabled = True),
    ]
  )]
)]

fig_dict = dict(data=data)

pio.show(fig_dict, validate=False)
```
