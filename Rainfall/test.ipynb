```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv("Rainfall_1901_2016_PAK.csv")
df.head()
```

```python
# EDA 
# Remove spaces and dashes 
df.rename(columns={'Rainfall - (MM)':'Rainfall(mm)',' Year':'year'},inplace=True)
df.columns
```

```python
plt.figure(figsize=(15,5))
plt.bar(df['year'], df['Rainfall(mm)'])
plt.xlabel('hour')
plt.ylabel('RainFall(mm)')
plt.title('Rainfall vs year', y=1.1)
plt.grid()
plt.show()
```

```python
df = df.groupby('year').mean()
```

```python
plt.figure(figsize=(15,5))
plt.bar(df.index, df['Rainfall(mm)'])
plt.xlabel('year')
plt.ylabel('Rainfall(mm) in Avg')
plt.title('Rainfall vs year', y=1.1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df['Rainfall(mm)'])
plt.xlabel('year')
plt.ylabel('Rainfall(mm) in Avg')
plt.title('Rainfall vs year', y=1.1)
plt.grid()
plt.show()
```

```python
import plotly.express as px
plt.figure(figsize=(50,50))
fig = px.line(df, x='year', y="Rainfall(mm)",line_group='year')
fig.add_hrect(y0="100", y1="160", 
              annotation_text="Flood Alarm", annotation_position="top left",
              fillcolor="green", opacity=0.20, line_width=1)
fig.update_layout(
    font_family="Courier New",
    font_color="red",font_size=15,title="Flood Indication with Respect to year",
    title_font_family="Times New Roman",
    title_font_color="Red", yaxis_title="Rainfall(mm)",
    legend_title_font_color="blue")

fig.show()
```
